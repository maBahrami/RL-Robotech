# Cart Pole env

```python
import gymnasium as gym
import time
import matplotlib.pyplot as plt
```

```python
env = gym.make("CartPole-v1", render_mode="rgb_array")
```

```python
obsv = env.reset()
print(obsv)
frame = env.render()
plt.imshow(frame)

while True:
    action = env.action_space.sample()
    next_state, reward, terminated, truncated, info = env.step(action)
    print(terminated)
    frame = env.render()
    plt.imshow(frame)
    time.sleep(0.1)

    if terminated == True:
        break
```
