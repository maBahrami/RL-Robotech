```python
env = gym.make("FrozenLake-v1", render_mode='ansi')
env.reset()

print(env.observation_space)
print(env.action_space)

print(env.unwrapped.P[1][2]) # transition probability
```

```python
print(env.step(2)) # doing action 2; we know that the env is stochastic
print(env.render())
```

```python
# executing a random action
env.reset()
random_action = env.action_space.sample()
next_state, reward, terminated, truncated, info = env.step(random_action)
print(next_state)
print(terminated)
print(env.render())
```

```python
episodes = 5

for i in range(episodes):
    print(f"--------- episode: {i+1} ------------")
    env.reset()
    print(env.render())
    while True:
        action = env.action_space.sample()
        next_step, reward, terminated, truncated, info = env.step(action)
        print(env.render())

        if terminated == True:
            break
```
